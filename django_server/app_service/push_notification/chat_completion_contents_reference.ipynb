```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import time

def crawl(url):
    options = Options()
    options.add_argument('-headless')

    driver = webdriver.Firefox(options=options)
    driver.get(url)

    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    ul_tag = soup.find('ul', {'class': 'wk_cont'})

    if ul_tag is not None:
        i_tags = ul_tag.find_all('i', {'class': 'prd_type t3'}, limit=2)

        for i_tag in i_tags:
            span = i_tag.find_next('span')
            if span is not None:
                print(span.text)

    driver.quit()
    
crawl('https://topis.seoul.go.kr/prdcBus/openPrdcBus.do')
```

```python
from dotenv import dotenv_values

config = dotenv_values(".env")
SEOUL_API_KEY = config["SEOUL_API_KEY"]
WEATHER_API_KEY = config["WEATHER_API_KEY"]
```

```python
import requests
import xml.etree.ElementTree as ET

axis_codes = [107, 12, 13, 133, 149, 173, 212, 219, 241, 250, 267,
             274, 309, 319, 327, 338, 342, 354, 365, 370, 377, 379,
              383, 386, 403, 42, 427, 450, 454, 6, 463, 47, 473, 
              481, 484, 57, 74, 89, 98]
link_ids = []

for axis_code in axis_codes:
    url = f"http://openapi.seoul.go.kr:8088/{SEOUL_API_KEY}/xml/LinkWithLoad/4/5/{axis_code}/"
    response = requests.get(url)

    root = ET.fromstring(response.content)

    for child in root:
        print(child.tag, child.text)
        if child.tag == 'row':
            for grandchild in child:
                print('\t', grandchild.tag, grandchild.text)
                if grandchild.tag == 'link_id':
                    link_ids.append(int(grandchild.text))
                
link_ids
```

```python
print(link_ids)
```

```python
speeds = []

for link_id in link_ids:
    url = f"http://openapi.seoul.go.kr:8088/{SEOUL_API_KEY}/xml/TrafficInfo/1/5/{link_id}/"
    response = requests.get(url)
    root = ET.fromstring(response.content)
    
    for row in root.iter('row'):
        prcs_spd = row.find('prcs_spd').text
        speeds.append(float(prcs_spd))
        print(f'Link ID: {link_id}, Speed: {prcs_spd}')
        
print(speeds)
```

```python
average_speed = sum(speeds) / len(speeds)
print(average_speed)
```

```python
import requests
import json

def get_weather(api_key, location_key):
    base_url = f"http://dataservice.accuweather.com/currentconditions/v1/{location_key}"
    params = {
        "apikey": api_key,
        "language": "en-us",
        "details": True
    }
    response = requests.get(base_url, params=params)
    return response.json()

def pretty_print(weather):
    print(f"WeatherText: {weather[0]['WeatherText']}")
    print(f"HasPrecipitation: {weather[0]['HasPrecipitation']}")
    print(f"Temperature: {weather[0]['Temperature']}")
    print(f"RealFeelTemperature: {weather[0]['RealFeelTemperature']}")
    print(f"Visibility: {weather[0]['Visibility']}")
    print(f"Wind Speed: {weather[0]['Wind']['Speed']}")
    print(f"RelativeHumidity: {weather[0]['RelativeHumidity']}")
    print(f"CloudCover: {weather[0]['CloudCover']}")
    print(f"PressureTendency: {weather[0]['PressureTendency']['LocalizedText']}")

api_key = WEATHER_API_KEY
location_key = '226081'
weather = get_weather(api_key, location_key)
pretty_print(weather)
```
